```python
features=['name', 'hobby', 'age','educational level', 'marital status']
sns.set_style("whitegrid");
sns.pairplot(df_with_columns,vars = features, hue = 'class', diag_kind = 'kde', 
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'}, size = 2) 
plt.show()
```
